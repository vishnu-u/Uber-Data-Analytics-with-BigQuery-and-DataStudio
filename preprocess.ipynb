```python
#Uber New York Analytics
import pandas as pd
import matplotlib.pyplot as plt
import datetime

#Datasets for the 6 month period is present in 6 different files
#Read them separately and combine them all to a single dataset
apr = pd.read_csv("Dataset/uber-raw-data-apr14.csv")
may = pd.read_csv("Dataset/uber-raw-data-may14.csv")
jun = pd.read_csv("Dataset/uber-raw-data-jun14.csv")
jul = pd.read_csv("Dataset/uber-raw-data-jul14.csv")
aug = pd.read_csv("Dataset/uber-raw-data-aug14.csv")
sep = pd.read_csv("Dataset/uber-raw-data-sep14.csv")

data = [apr,may,jun,jul,aug,sep]
data = pd.concat(data)
data
```

```python
mm=[]
dd=[]
yy=[]
hrs=[]
dow=[]
dow_list = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]

dt_time = data['Date/Time'].copy(deep=True).to_numpy()
for i in range(len(dt_time)):
    mm.append(dt_time[i].split(" ")[0].split("/")[0])

dt_time = data['Date/Time'].copy(deep=True).to_numpy()
for i in range(len(dt_time)):
    dd.append(dt_time[i].split(" ")[0].split("/")[1])

dt_time = data['Date/Time'].copy(deep=True).to_numpy()
for i in range(len(dt_time)):
    yy.append(dt_time[i].split(" ")[0].split("/")[2])

dt_time = data['Date/Time'].copy(deep=True).to_numpy()
for i in range(len(dt_time)):
    hrs.append(dt_time[i].split(" ")[1].split(":")[0])

for i in range(len(mm)):
    dow.append(dow_list[ datetime.date( int(yy[i]),
                                        int(mm[i]),
                                        int(dd[i])).weekday() 
                        ])
    

data['Mm'] = mm
data['Dd'] = dd
data['Yy'] = yy 
data['Hrs'] = hrs
data['Dw'] = dow
del data['Date/Time']
print(data)

data.to_csv("Dataset/uber.csv")
```
